There are two purposes in this project: 1. Find faces in a group photo; 2. Crop face images from a group photo and save them in a subdicectory. Note: Before running, please check there is no subdirectory in the folder. If any subdirectory exists, please delete it first.

```python
# In this cell, some libraries were imported
import cv2
import sys
import os
from PIL import Image, ImageDraw
import pylab
```

```python
# Face Detection Function
def detectFaces(image_name):
    print ("Face Detection Start.")
    # Read the image and convert to gray to reduce the data
    img = cv2.imread(image_name)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)#Color => Gray
    
    # The haarcascades classifier is used to train data
    face_cascade = cv2.CascadeClassifier("/usr/share/opencv/haarcascades/haarcascade_frontalface_default.xml")
    faces = face_cascade.detectMultiScale(gray, 1.2, 5)#1.3 and 5are the min and max windows of the treatures
    result = []
    for (x,y,width,height) in faces:
        result.append((x,y,x+width,y+height))
    print ("Face Detection Complete.")
    return result
```

```python
# Fuction to crop and save Face images
def saveFaces(image_name):
    print ("Face Images-crop Start.")
    faces = detectFaces(image_name)
    if faces:
        #All croped face images will be saved in a subdirectory
        save_dir = image_name.split('.')[0]+"_faces"
        os.mkdir(save_dir)
        count = 0
        print ("All cropped face images are coming from Function 'saveFaces'.")
        for (x1,y1,x2,y2) in faces:
            file_name = os.path.join(save_dir,str(count)+".jpg")
            Image.open(image_name).crop((x1,y1,x2,y2)).save(file_name)

            img = Image.open(file_name)
            axis('off')
            imshow(img)
            show()
            count+=1
            print ("     Face No." + str(count) + '.')
    print ("The total number of faces in this picture are " + str(count) + '.')
    print ("Face Images-crop Complete.")
    return count
```

```python
# Fuction to rectangle all faces in a group images
def drawFaces(image_name):
    print ("Face Rectangled Start.")
    faces = detectFaces(image_name)
    if faces:
        img = Image.open(image_name)
        draw_instance = ImageDraw.Draw(img)
        for (x1,y1,x2,y2) in faces:
            draw_instance.rectangle((x1,y1,x2,y2), outline=(255, 0,0))
        img.save('new'+image_name) 
    print ("Face Rectangled Complete.")
```

```python
import time
# To call all functions
image_name = '01.jpg'
start1 = time.time()
count = saveFaces(image_name)
end1 = time.time()
time_saveFace = end1 - start1
print ("The total time-consuming of saceFace function is " + str(time_saveFace)+ 's')
print ("The average time-consuming of saceFace function is " + str(time_saveFace / count)+ 's')

start2 = time.time()
drawFaces(image_name)
end2 = time.time()
time_drawFaces = end2 - start2
print ("The time-consuming of drawFaces function is " + str(time_drawFaces)+ 's')
```
